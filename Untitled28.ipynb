```python
f = open('wikipedia_ai.txt', 'r', encoding='UTF8')
text = f.read()
print(text)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wc_output = WordCloud(width = 1500, height = 1000).generate(text)

plt.figure( figsize=(12, 8))

plt.imshow(wc_output)
plt.axis('off')
plt.show()
```
