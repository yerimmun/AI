```python
import matplotlib.pyplot as plt
import numpy as np

xlist = np.arange(-3.14, 3.14*2, 0.3)
y_sin = np.sin(xlist)
y_cos = np.cos(xlist
               )
plt.plot(xlist, y_sin)
plt.plot(xlist, y_cos)

plt.title('sine & cosine graph')
plt.xlabel('x value')
plt.ylabel('y value')
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label = 'sine')
plt.plot(x, y2, label = 'cos')

plt.title('sin & cos')
plt.xlabel('x value')
plt.ylabel('y value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

scores = []
for i in range(5):
    value = int(input(' '))
    scores.append(value)

sumR = np.sum(scores)
aveR = np.mean(scores)
varR = np.var(scores)
stdR = np.std(scores)

print('', sumR)
print('', aveR)
print('', varR)
print('', stdR)
```
